```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create the new dataset
data = {
    'Number of Products Purchased': [3, 5, 2, 4, 6, 1, 4, 2, 5],
    'Monthly Expenditure': [1500, 7000, 4000, 6000, 8000, 300, 7500, 4500, 6500],
    'Average Transaction Amount': [20000, 30000, 15000, 25000, 40000, 10000, 35000, 18000, 32000]

}
df = pd.DataFrame(data)
print(df)
```

```python
# Combine the features (Number of Products Purchased, Monthly Expenditure, Average salary)
X = df[['Number of Products Purchased', 'Monthly Expenditure', 'Average Transaction Amount']].values
print(X)
```

Euclidean Distance between three points P(x1, y1, z1) and Q(x2, y2, z2):
Distance(P, Q) = sqrt((x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2).

```python
# Apply K-means clustering with 2 clusters
kmeans = KMeans(n_clusters=2)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Plot the data and visualize the clusters
fig = plt.figure(figsize=(12, 6))

ax2 = fig.add_subplot(122)
ax2.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
ax2.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
ax2.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='green', label='Centroids')
ax2.set_title('K-means Clustering (2D Projection)')
ax2.set_xlabel('Number of Products Purchased')
ax2.set_ylabel('Monthly Expenditure')
ax2.legend()

plt.show()
```

# **Same problem but finding 2 clusters while combining two different features (Number of Products Purchased,Average Transaction Amount), and (Monthly Expenditure, Average Transaction Amount).**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create the new dataset
data = {
    'Number of Products Purchased': [3, 5, 2, 4, 6, 1, 4, 2, 5],
    'Monthly Expenditure': [1500, 7000, 4000, 6000, 8000, 300, 7500, 4500, 6500],
    'Average Transaction Amount': [20000, 30000, 15000, 25000, 40000, 10000, 35000, 18000, 32000]
}
df = pd.DataFrame(data)
```

```python
# Feature 1: (Number of Products Purchased, Average Transaction Amount)
X1 = df[['Number of Products Purchased', 'Average Transaction Amount']].values

# Feature 2: (Monthly Expenditure, Average Transaction Amount)
X2 = df[['Monthly Expenditure', 'Average Transaction Amount']].values

# Apply K-means clustering with 2 clusters for Feature 1
kmeans1 = KMeans(n_clusters=2, random_state=42)
y_kmeans1 = kmeans1.fit_predict(X1)

# Apply K-means clustering with 2 clusters for Feature 2
kmeans2 = KMeans(n_clusters=2, random_state=42)
y_kmeans2 = kmeans2.fit_predict(X2)
```

```python
# Plot the data and visualize the clusters for Feature 1
plt.subplot(121)
plt.scatter(X1[y_kmeans1 == 0, 0], X1[y_kmeans1 == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X1[y_kmeans1 == 1, 0], X1[y_kmeans1 == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(kmeans1.cluster_centers_[:, 0], kmeans1.cluster_centers_[:, 1], s=300, c='green', label='Centroids')
plt.xlabel('Number of Products Purchased')
plt.ylabel('Average Transaction Amount')
plt.legend()

# Plot the data and visualize the clusters for Feature 2
plt.subplot(122)
plt.scatter(X2[y_kmeans2 == 0, 0], X2[y_kmeans2 == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X2[y_kmeans2 == 1, 0], X2[y_kmeans2 == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], s=300, c='green', label='Centroids')

plt.xlabel('Monthly Expenditure')
plt.ylabel('Average Transaction Amount')
plt.legend()

plt.tight_layout()
plt.show()
```
