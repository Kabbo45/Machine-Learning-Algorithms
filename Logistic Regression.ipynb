In this table, Xi represents the input feature values and Yi represents the corresponding target output values. Each row represents a data point in the dataset. Apply Logistic Regression (LoR) to the above dataset and write a Python program to illustrate the whole implementation procedure. Also, predict what will be the outcome of Predict (0.5), and Predict (1.7) based on the LoR model you developed.


```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Define the dataset
X = np.array([[0.2], [0.3], [0.5], [0.8], [1.1], [1.3], [1.5], [1.9], [2.4], [2.7]])
y = np.array([0, 0, 0,0, 1, 0, 1, 1, 1, 1])

# Create the logistic regression model and fit it to the data
lr = LogisticRegression()
lr.fit(X, y)

# Predict the outcomes for the values of 0.35 and 0.47
X_new = np.array([[0.5], [1.7]])
y_pred = lr.predict(X_new)
print("Predictions for 0.5 and 1.7:", y_pred)
```
