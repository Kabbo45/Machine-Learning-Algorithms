In this program, we define the dataset where the input features X are the exam scores (in this case, represented as percentages) and the target outputs y are the corresponding SGPA values.
| Exam Score (X1) | Study Hours (X2) | SGPA |
|---------------|-----------------|------|
| 50            | 10              | 3.1  |
| 60            | 12              | 3.4  |
| 70            | 15              | 3.2  |
| 80            | 18              | 3.6  |
| 90            | 20              | 3.65 |

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Define the dataset
X = np.array([[50, 60, 70, 80, 90], [10, 12, 15, 18, 20]]).T
y = np.array([3.1, 3.4, 3.2, 3.6, 3.65])

# Create the linear regression model and fit it to the data
regression = LinearRegression()
regression.fit(X, y)

# Predict the SGPA for new exam scores (0.35 and 0.47)
X_new = np.array([[55], [10]]).T
sgpa_pred = regression.predict(X_new)
print("Estimated SGPA for 55 exam score and 10 hours study:", sgpa_pred)
```
