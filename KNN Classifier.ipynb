| 1          | 4        | good   |
In this table, each row represents a data point. The "Durability" column represents the durability value, the "Strength" column represents the strength value, and the "Result" column represents the corresponding classification result ("bad" or "good").

```python
from sklearn.neighbors import KNeighborsClassifier

# Define the dataset
Durability = [7,  7, 3, 1 ]
Strength = [7, 4, 4, 4 ]
Result = ["bad", "bad", "good", "good"]

# Create the feature matrix X
X = list(zip(Durability, Strength))

# Create the target vector y
y = Result

# Create the KNN classifier and fit the data
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X, y)

# Predict the class for a new data point
new_data = [(5, 3)]
predicted_class = knn.predict(new_data)
print("Predicted class:", predicted_class)
```
